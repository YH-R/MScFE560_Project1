```python
import pandas as pd
import numpy as np
```

```python
## Reading csv
data = pd.read_csv('MScFE560_FM_GWP1_Data.csv')
avg_returns = data[:1]
std_dev = data[1:2]
cov_matrix = data[6:]
std_dev
```

```python
cov_matrix['rows'] = ['AAPL','AMZN','NFLX','META','GOOG']
cov_matrix
```

## Analysis for Portfolio A

```python
stocks = ['AAPL','AMZN','NFLX','META','GOOG']
```

```python
for i in stocks:
    for j in stocks:
        if(i==j):
            continue
        best_coeff_var = 10**8
        best_wa = 0
        best_wb = 0
        best_std = 0
        best_ret = 0
        for wa in np.arange(-2,1,0.01):
            wb = 1 - wa
            if(max(wa,wb)<=0):
                continue
            expected_return = wa * float(avg_returns[i]) + wb * float(avg_returns[j])
            if(expected_return<=0):
                continue
            prt_std_dev = (wa**2 * float(std_dev[i])**2 + wb**2 * float(std_dev[j])**2 + 2*wa*wb*float(std_dev[i])*float(std_dev[j])*float(cov_matrix[cov_matrix.rows==i][j]))**0.5
            coeff_variation = prt_std_dev / expected_return
            if(coeff_variation < best_coeff_var):
                best_coeff_var = coeff_variation
                best_wa = wa
                best_wb = wb
                best_ret = expected_return
                best_std = prt_std_dev
        print(f"With wa = {best_wa} and wb = {best_wb} weights, with expected returns {best_ret} and with std_dev {best_std}  has the best coeff_var for {i} and {j} and is {best_coeff_var}")
    
```

# Portfolio C

```python
for i in stocks:
    for j in stocks:
        for k in stocks:
            if(i==j or j==k or i==k):
                continue
            best_coeff_var = 10**8
            best_wa = 0
            best_wb = 0
            best_wc = 0
            best_dev = 0
            best_ret = 0
            for wa in np.arange(-2,1,0.01):
                for wb in np.arange(-2,1,0.01):    
                    wc = 1 - wa - wb
                    if(max(wa,wb,wc)<=0):
                        continue
                    expected_return = wa * float(avg_returns[i]) + \
                                      wb * float(avg_returns[j]) + \
                                      wc * float(avg_returns[k])

                    if(expected_return<=0):
                        continue
                    prt_std_dev = (wa**2 * float(std_dev[i])**2 + 
                                   wb**2 * float(std_dev[j])**2 +
                                   wc**2 * float(std_dev[k])**2 + 
                                   2*wa*wb*float(std_dev[i])*float(std_dev[j])*float(cov_matrix[cov_matrix.rows==i][j]) + 
                                   2*wa*wc*float(std_dev[i])*float(std_dev[k])*float(cov_matrix[cov_matrix.rows==i][k]) +
                                   2*wc*wb*float(std_dev[k])*float(std_dev[j])*float(cov_matrix[cov_matrix.rows==k][j])
                                   )**0.5
                    coeff_variation = prt_std_dev / expected_return
                    if(coeff_variation < best_coeff_var):
                        best_coeff_var = coeff_variation
                        best_wa = wa
                        best_wb = wb
                        best_wc = wc
                        best_dev= prt_std_dev
                        best_ret = expected_return
            print(f"With wa = {best_wa}, wb = {best_wb} and wc = {best_wc} weights,  with expected returns {best_ret} and with std_dev {best_dev} has the best coeff_var for {i},{j} and {k}, and is {best_coeff_var}")
    
```
